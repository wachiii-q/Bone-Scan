```python
# merge  extracted information

import pandas as pd
from bonescan.utils.common import log
```

```python
mainReportPath = "data/2023/mapping.csv"
snrReportPath = "data/2023/SNR_CNR/snr_cnr_results.csv"

mainReportDf = pd.read_csv(mainReportPath)
snrReportDf = pd.read_csv(snrReportPath)

log(mainReportDf.columns)
log(snrReportDf.columns)
```

```python
mergedDf = pd.merge(mainReportDf, snrReportDf, on="ACC")
```

```python
log(mergedDf.columns)
```

```python
log(mergedDf.head())
```

```python
# drop nan values
mergedDf = mergedDf.dropna()
```

```python
# save the merged data
mergedDf.to_csv("merged_snr_2023.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data1 = {'Product': ['A', 'B', 'C'], 'Sales_2023': [100, 150, 80]}
data2 = {'Product': ['A', 'B', 'C'], 'Sales_2024': [120, 180, 90]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Create the base bar plot
plt.bar(df1.index, df1['Sales_2023'], width=0.6)

# Create the overlapping bars
plt.bar(df2.index + 0.2, df2['Sales_2024'], width=0.6, alpha=0.7)  # Adjust alpha for transparency

plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(df1.index + 0.2, df1['Product'])
plt.legend(['2023', '2024'])
plt.show()
```
